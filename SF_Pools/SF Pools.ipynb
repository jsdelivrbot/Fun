# Mapping Location and Info on SF Pools

Plots map with all the pool locations color coded by type: SF Recreation, YMCA, University, or OWS.

Provides URL on popup for each location.

### Boring Declarations

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from IPython.display import display
import seaborn as sns
import matplotlib
import matplotlib.pylab as plt
pd.set_option('precision',1) #easy to read output

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
import folium
```

### Building Pool Dataset

```python
import folium
# Pool Locations
colors={'Uni':'red',
        'YMCA':'blue',
        'SFrec':'green',
        'OWS':'yellow'}

PoolDict ={'USF Koret Health & Recreation': [37.777135, -122.454494,'Uni','https://www.usfca.edu/koret/aquatics'],
             'UCSF Baker Rec Center': [37.768534, -122.393288,'Uni','https://campuslifeservices.ucsf.edu/fitnessrecreation/services/schedules'],
             'SF Rec North Beach':[37.8027052, -122.4124234,'SFrec','http://sfrecpark.org/destination/joe-dimaggio-playground/north-beach-pool/'],
             'SF Rec Garfield Pool': [37.7500745, -122.4120347,'SFrec','http://sfrecpark.org/destination/garfield-square/garfield-pool/'],
             'SF Martin Luther King Pool': [37.7254915, -122.3935752,'SFrec','http://sfrecpark.org/destination/bay-view-playground/martin-luther-king-jr-pool/'],
             'SF Rossi Pool': [37.7790784,-122.4582621,'SFrec','http://sfrecpark.org/destination/angelo-j-rossi-playground/rossi-community-pool/'],
             'SF Sava Pool': [37.7377925,-122.4758529,'SFrec','http://sfrecpark.org/destination/carl-larsen-park/sava-pool/'],
             'SF Hamilton Pool': [37.7845604, -122.4354450,'SFrec','http://sfrecpark.org/destination/hamilton-rec-center/hamilton-pool/'],
             'SF Balboa Pool': [37.7261962, -122.4432667,'SFrec','http://sfrecpark.org/destination/balboa-park/balboa-pool/'],
             'SF Mission Pool': [37.7594975,-122.4226561,'SFrec','http://sfrecpark.org/destination/mission-playground/mission-community-pool/'],
             'YMCA Presidio': [37.800198, -122.454897,'YMCA','https://www.ymcasf.org/locations/presidio-community-ymca'],
             'YMCA Embarcadero': [37.792921, -122.392000,'YMCA','https://www.ymcasf.org/locations/embarcadero-ymca'],
             'YMCA China Town': [37.793486, -122.406977,'YMCA','https://www.ymcasf.org/locations/chinatown-ymca'],
             'Aquatics Park OWS':[37.807028, -122.423396,'OWS','http://www.ggtc.org/content.aspx?page_id=22&club_id=424851&module_id=102385']
            }
PoolDF = pd.DataFrame.from_dict(PoolDict,orient='index')

#PoolDF.head(15)
```

### Mapping Pool Locations

```python
# Make a map of the pool locations
pools = PoolDF.index.values
fmap  = folium.Map(location=[37.76,-122.45],tiles='openstreetmap',zoom_start=12,width=600,height=500)
for pool in pools:
    poolarr = PoolDF.loc[pool]
    popup = ("{name}<br>"
    "{url}<br>"
   ).format(name=pool,
            url='<a href="{}">URL</a>'.format(poolarr[3])
            )
    folium.Marker(poolarr[0:2],
                  popup=popup,
                  icon = folium.Icon(color=colors[poolarr[2]])).add_to(fmap)
fmap.save('PoolMap.html')
fmap
```
